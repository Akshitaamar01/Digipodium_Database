```sql
use practice;
```

```sql
-- Q1 Simple SELECT Retrieve all columns from the "houseprices" table.
select * from houseprices
```

```sql
-- Q2. Filtering with WHERE Retrieve the names of players from the "houseprices" table who have scored more than 100 runs.
select name, runs from houseprices where runs > 100;
```

```sql
--Q3. Aggregation with GROUP BY Calculate the total number of games played by each player.
SELECT name, COUNT(games) AS total_games_played
FROM houseprices
GROUP BY name;
```

```sql
-- Q4 Joining Tables Retrieve the player names and their corresponding teams from the "houseprices" table.
select n.name,t.team
from houseprices n
join on n.name = t.name;
```

```sql
-- Q5 ORDER BY Retrieve player names and their number of hits, ordered by the number of hits in descending order.
select name, hits 
from houseprices
order by hits desc;
```

```sql
-- Q6. LIMIT Retrieve the top 10 players with the highest number of home runs
select name, runs 
from houseprices
order by runs desc
limit 10;
```

```sql
-- Q7. DISTINCT Retrieve unique team names from the "houseprices" table.
Select distinct name as 'unique names'
from houseprices;
```

```sql
-- Q8 Subqueries Retrieve the names of players who have more home runs than the average number of home runs.
select name
from houseprices
where runs > (SELECT AVG(runs) FROM houseprices);
```

```sql
-- Q9  Updating Data Update the team name of the player with Num 2 to 'CHICAGO'.
UPDATE houseprices
SET team = 'CHICAGO'
WHERE Num = 2;

select * from houseprices;
```

```sql
-- Q10.  Deleting Data Delete the player with Num 3 from the table.
delete from houseprices
where num =3;

Select * from houseprices;
```

```sql
-- Q11. COUNT with WHERE Count the number of players who have played more than 150 games.
SELECT COUNT(*) AS num_players
FROM houseprices
WHERE games > 150;
```

```sql
-- Q12. Nested SELECT Retrieve the names of players who have scored the most runs.
SELECT name
FROM houseprices
WHERE runs = (
    SELECT MAX(runs)
    FROM houseprices
);
```

```sql
-- Q13. Updating Data Based on a Condition Increase the number of RBIs by 10 for all players who have more than 20 home runs.
UPDATE houseprices
SET RBIs = RBIs + 10
WHERE runs > 20;

select * from houseprices
```

```sql
-- Q14. Calculated Columns Retrieve the total number of hits per game for each player.
SELECT name, hits, games, (hits / games) AS hits_per_game
FROM houseprices;
```

```sql
-- Question 15: NULL Values Retrieve the players with NULL values in the "Doubles" column.
SELECT name
FROM houseprices
WHERE Doubles IS NULL;
```

```sql
-- Q16. Aggregation with HAVING Retrieve the teams and their total number of games played, but only for teams with more than 100 games played.
SELECT name, SUM(games) AS total_games_played
FROM houseprices
GROUP BY team
HAVING total_games_played > 100;
```

```sql
-- Q17. Multiple Conditions with WHERE Retrieve the names of players who have more than 30 home runs and more than 50 RBIs.
SELECT name
FROM houseprices
WHERE runs > 30 AND RBIs > 50;
```

```sql
-- Q18. LIKE Operator Retrieve the names of players whose names start with 'A'.
SELECT name
FROM houseprices
WHERE name LIKE 'A%';
```

```sql
-- Q19. SUM with GROUP BY Calculate the total number of runs scored by each player.
SELECT name, SUM(runs) AS total_runs_scored
FROM houseprices
GROUP BY name;
```

```sql
-- Q20. Conditional ORDER BY Retrieve player names and their number of hits, ordered by the number of hits in descending order. If hits are equal, order them by runs in ascending order.
SELECT name, hits, runs
FROM houseprices
ORDER BY hits DESC, runs ASC;
```
