```sql
use practice;
```

```sql
select * from baseball;
```

```sql
-- Q1. Retrieve the total number of games played by each player.
SELECT name, SUM(games) AS total_games_played
from baseball
GROUP BY name;
```

```sql
-- Q2. List the names and total hits for players who have scored more than 100 runs.
select name, hits 
from baseball
where runs > 100;
```

```sql
-- Q3. Display the player with the highest number of home runs.
select name, runs
from baseball
order by runs desc
limit 1;
```

```sql
-- Q4. Show the average number of runs scored by players for each team.
SELECT name, AVG(runs) AS average_runs
FROM baseball
GROUP BY name;
```

```sql
-- Q5. Find the player with the maximum at-bats.
select name, AtBats
from baseball
order by atbats DESC
limit 1;
```

```sql
-- Q6. List the names of players with more than 30 doubles.
select name, doubles
from baseball
where doubles > 30;
```

```sql
-- Q7. Show the player with the most triples.
select name, triples
from baseball
order by triples DESC
limit 1;
```

```sql
-- Q8. Display the names of players who have hit at least 20 home runs and have more than 80 RBIs.
select name, runs, RBIs
from baseball
where runs >= 20 and RBIs > 80;
```

```sql
-- Q9. Find the player with the highest batting average (hits divided by at-bats).
SELECT name, (hits / atbats) AS batting_average
FROM baseball
ORDER BY batting_average DESC
LIMIT 1;
```

```sql
-- Q10. List the players who have played fewer than 100 games.
select name, games
from baseball
where games < 100;
```

```sql
-- Q11. Display the player with the maximum and minimum number of RBIs.
select name, RBIs
from baseball
WHERE RBIs = (SELECT MAX(RBIs) FROM baseball)
UNION
SELECT name, RBIs
FROM baseball
WHERE RBIs = (SELECT MIN(RBIs) FROM baseball);
```

```sql
-- Q12. Show the names of players who have played exactly 150 games.
select name, games
from baseball
where games = 150;
```

```sql
-- Q13. Find the player with the highest number of runs per game (runs divided by games).
select name, (runs/games) as Runs_per_game
from baseball
order by runs_per_game DESC
limit 1;
```

```sql
-- Q14. List the players who have hit at least 10 home runs and played for 'NYY' or 'BOS'.
SELECT name, team
FROM baseball
WHERE runs >= 10
AND (team = 'NYY' OR team = 'BOS');
```

```sql
-- Q15. Display the player with the highest total runs and the lowest total hits.
select name, runs, hits
from baseball
where runs = (select max(runs) from baseball)
And hits = (select min(hits) from baseball);
```

```sql
-- Q16. Show the players with more than 10 doubles and a batting average greater than 0.300
SELECT name, doubles
FROM baseball
WHERE doubles > 10 AND (CAST(hits AS DECIMAL) / atbats) > 0.300;
```

```sql
-- Q17. List the players who have hit a home run but have less than 50 runs scored.
SELECT name, runs
FROM baseball
WHERE runs > 0 AND runs < 50;
```

```sql
-- Q18. Display the names of players who have at least 40 hits and have played fewer than 50 games.
SELECT name, hits, games
FROM baseball
WHERE hits >= 40 AND games < 50;
```

```sql
-- Q19. Find the player with the highest batting average among those with at least 300 at-bats.
SELECT name, batting_average
FROM baseball
WHERE atbats >= 300
ORDER BY batting_average DESC
LIMIT 1;
```

```sql
-- Q20. List the players with more than 25 doubles and an average of at least 3 RBIs per game.
SELECT name
FROM baseball
WHERE doubles > 25 AND (RBIs / games) >= 3;
```
